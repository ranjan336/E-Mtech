```python
import os
import json
import pandas as pd
```

```python

json_file_path = r"C:\Users\ranjan.patra\OneDrive - Lingaro Sp. z o. o\DATA\IITJ\Course\Projects\ICC\IPL"
files=os.listdir(json_file_path)
info_list = []
```

```python
# Iterate through each JSON file
for file_name in files:
    # Check if the file is a JSON file
    if file_name.endswith(".json"):
        # Construct the full file path
        file_path = os.path.join(json_file_path, file_name)

        # Open and read the JSON file
        try:
            with open(file_path, "r") as json_file:
                data = json.load(json_file)

            # Extract information from the JSON file and add it to the info_list
            info = {
                'match_id': file_name,
                'Venue': data["info"]["venue"],
                'Date': data["info"]["dates"][0],
                'Team_1': data["info"]["teams"][0],
                'Team_2': data["info"]["teams"][1],
                'Toss_decision': data["info"]["toss"]["decision"],
                'Toss_Winner': data["info"]["toss"]["winner"],
                'Match_outcome': data["info"]["outcome"],
                'Match_Result_Wicket': data["info"]["outcome"].get("by",{}).get("wickets",None),
                'Match_Result_Run': data["info"]["outcome"].get("by",{}).get("runs",None),
                'Match_Winner': data["info"]["outcome"].get("winner",None)
            }
            
            info_list.append(info)

        except json.JSONDecodeError as e:
            # Handle the JSON schema mismatch error (or any other error) here
            print(f"Error reading {file_name}: {e}")
            
        except Exception as e:
            # Handle other exceptions here, if necessary
            print(f"An error occurred while processing {file_name}: {e}")
```

```python
df_info = pd.DataFrame(info_list)
display(df_info)
```

```python
import pandas as pd

# Assuming you have DataFrames named 'df_match' and 'df_info_clean'

# Create an empty DataFrame to store the results
results_df = pd.DataFrame(columns=[
    "Team_1", "Team_2", "Total_Matches", "Total_Wins_Team_1", "Total_Wins_Team_2",
    "Total_Toss_Wins_Team_1", "Total_Toss_Wins_Team_2", "Match_Toss_Wins_Team_1", "Match_Toss_Wins_Team_2",
    "Match_Win_Toss_Loss_Team_1", "Match_Win_Toss_Loss_Team_2"
])

# Iterate through unique team combinations in df_match
for i in range(len(df_match)):
    team_1, team_2 = df_match.iloc[i, 0], df_match.iloc[i, 1]

    # Filter df_info_clean for the current team combination
    filtered_data = df_info_clean[
        ((df_info_clean['Team_1'] == team_1) & (df_info_clean['Team_2'] == team_2)
        | (df_info_clean['Team_1'] == team_2) & (df_info_clean['Team_2'] == team_1))
        & ((df_info_clean['Toss_Winner'] == team_1) | (df_info_clean['Toss_Winner'] == team_2))
    ]

    total_matches = len(filtered_data)
    total_wins_team_1 = len(filtered_data[filtered_data['Match_Winner'] == team_1])
    total_wins_team_2 = len(filtered_data[filtered_data['Match_Winner'] == team_2])
    total_toss_wins_team_1 = len(filtered_data[filtered_data['Toss_Winner'] == team_1])
    total_toss_wins_team_2 = len(filtered_data[filtered_data['Toss_Winner'] == team_2])
    match_toss_wins_team_1 = len(filtered_data[
        (filtered_data['Match_Winner'] == team_1) & (filtered_data['Toss_Winner'] == team_1)
    ])
    match_toss_wins_team_2 = len(filtered_data[
        (filtered_data['Match_Winner'] == team_2) & (filtered_data['Toss_Winner'] == team_2)
    ])
    match_win_toss_loss_team_1 = len(filtered_data[
        (filtered_data['Match_Winner'] == team_1) & (filtered_data['Toss_Winner'] != team_1)
    ])
    match_win_toss_loss_team_2 = len(filtered_data[
        (filtered_data['Match_Winner'] == team_2) & (filtered_data['Toss_Winner'] != team_2)
    ])

    # Add the results to the results DataFrame
    results_df = results_df.append({
        "Team_1": team_1,
        "Team_2": team_2,
        "Total_Matches": total_matches,
        "Total_Wins_Team_1": total_wins_team_1,
        "Total_Wins_Team_2": total_wins_team_2,
        "Total_Toss_Wins_Team_1": total_toss_wins_team_1,
        "Total_Toss_Wins_Team_2": total_toss_wins_team_2,
        "Match_Toss_Wins_Team_1": match_toss_wins_team_1,
        "Match_Toss_Wins_Team_2": match_toss_wins_team_2,
        "Match_Win_Toss_Loss_Team_1": match_win_toss_loss_team_1,
        "Match_Win_Toss_Loss_Team_2": match_win_toss_loss_team_2
    }, ignore_index=True)

# Display the results DataFrame

```
